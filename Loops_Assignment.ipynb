## Student Marks Entry

### Goal:
    Record student marks and skip invalid entries.

### Steps:

    Create an empty list marks = [].
    
    Ask user to enter marks for 5 students.
    
    If marks < 0 or > 100 ‚Üí print ‚ÄúInvalid marks‚Äù and continue.
    
    If user enters -1 ‚Üí stop taking input (break).
    
    Print valid marks and their average.

### Example:

    Enter marks: 90  
    Enter marks: 105 ‚Üí Invalid marks  
    Enter marks: 80  
    Enter marks: -1 ‚Üí stop  
    Valid Marks: [90, 80]  
    Average = 85.0

```python
marks = []

while len(marks) < 5:  
    try:
        mark = int(input("Enter marks: "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if mark == -1:
        break

    if mark < 0 or mark > 100:
        print("Invalid marks")
        continue

    marks.append(mark)

print("Valid Marks:", marks)

if marks:
    average = sum(marks) / len(marks)
    print("Average =", average)
else:
    print("No valid marks entered.")
```

## üè® 8. Hotel Room Booking

### Goal:
    Manage simple room booking.

### Task:

    Create a list of rooms [101, 102, 103, 104, 105]
    
    Ask user to choose a room.
    
    If room already booked, print ‚ÄúRoom not available.‚Äù
    
    Continue until all rooms are booked.

### Example:

    Available rooms: [101, 102, 103, 104, 105]
    Choose room: 103
    Room 103 booked successfully!

```python
rooms = [101, 102, 103, 104, 105]

while rooms:
    print("Available rooms:", rooms)
    
    try:
        room_choice = int(input("Choose room: "))
    except ValueError:
        print("Please enter a valid room number.")
        continue

    if room_choice in rooms:
        print(f"Room {room_choice} booked successfully!")
        rooms.remove(room_choice)
    else:
        print("Room not available.")

print("All rooms are booked!")
```

## Car Rental System

### Goal:
    Create a car rental booking system.

### Steps:

    Cars = ["Civic", "Corolla", "City", "Alto"]
    
    Use a for loop to let multiple users book.
    
    If user enters "stop" ‚Üí break.
    
    If car not in list ‚Üí continue (invalid).
    
    Add valid bookings to booked_cars list.

### Example:

    Enter car name: Civic  
    Enter car name: Alto  
    Enter car name: Tesla ‚Üí Not available  
    Enter car name: stop  
    Booked Cars: ['Civic', 'Alto']

```python
cars = ["Civic", "Corolla", "City", "Alto"]
booked_cars = []

for i in range(100): 
    car_choice = input("Enter car name or stop: ").strip()
    
    if car_choice.lower() == "stop":
        break
    
    matched_car = None
    for car in cars:
        if car_choice.lower() == car.lower():
            matched_car = car
            break
    
    if not matched_car:
        print("Not available")
        continue
    
    booked_cars.append(matched_car)

print("Booked Cars:", booked_cars)
```

## School Marks Report
### Goal:

    Take input for multiple students‚Äô marks in 3 subjects and calculate their total and average.

### Steps:

    Outer loop ‚Üí for each student (3 students).
    
    Inner loop ‚Üí take marks for 3 subjects.
    
    Store each student‚Äôs total in a list.
    
    Print all totals and class average.

### Example:
    Enter marks for Student 1:
    Subject 1: 80
    Subject 2: 75
    Subject 3: 90
    Enter marks for Student 2:
    Subject 1: 60
    Subject 2: 70
    Subject 3: 65
    ‚Üí Output: [245, 195]

```python
student_totals = []

for student in range(1, 4):
    print(f"Enter marks for Student {student}:")
    total = 0
    for subject in range(1, 4):
        while True:
            try:
                mark = int(input(f"Subject {subject}: "))
                if 0 <= mark <= 100:
                    break
                else:
                    print("Enter marks between 0 and 100.")
            except ValueError:
                print("Invalid input. Enter a number.")
        total += mark
    student_totals.append(total)

print("Total marks of students:", student_totals)

class_average = sum(student_totals) / len(student_totals) if student_totals else 0
print("Class Average:", class_average)
```

## Shop Sales Tracker
### Goal:

    Track daily sales of 3 products over 5 days and calculate the total revenue.

### Steps:

    Outer loop ‚Üí iterate over 5 days.
    
    Inner loop ‚Üí take sales amount for 3 products each day.
    
    Store daily total in a list.
    
    If any product sale = 0 ‚Üí continue (skip that product).
    
    Print all daily totals and grand total.

### Example:
    Day 1:
    Product 1: 200
    Product 2: 150
    Product 3: 0 ‚Üí skipped
    Day 2:
    Product 1: 300
    Product 2: 250
    Product 3: 400
    ‚Üí Daily Totals = [350, 950], Grand Total = 1300

```python
daily_totals = []

for day in range(1, 6):
    print(f"Day {day}:")
    daily_total = 0
    for product in range(1, 4):
        while True:
            try:
                sale = int(input(f"Product {product}: "))
                break
            except ValueError:
                print("Invalid input. Enter a number.")
        
        if sale == 0:
            continue
        
        daily_total += sale
    daily_totals.append(daily_total)

grand_total = sum(daily_totals)

print("Daily Totals =", daily_totals)
print("Grand Total =", grand_total)
```

## Cinema Seat Booking System
### Goal:

    Take user input for seat availability across rows and columns, and mark booked seats.

### Steps:

    Outer loop ‚Üí for each row (3 rows).
    
    Inner loop ‚Üí for 5 seats in that row.
    
    Ask if seat is booked (Y/N).
    
    If booked ‚Üí add ‚ÄúBooked‚Äù else ‚ÄúAvailable‚Äù.
    
    Store in nested list and print seating chart.

### Example:
    Row 1:
    Seat 1 (Y/N): Y
    Seat 2 (Y/N): N
    Seat 3 (Y/N): Y
    Seat 4 (Y/N): N
    Seat 5 (Y/N): N

## Hotel Room Booking Tracker
### Goal:

    Record bookings for rooms on each floor.

### Steps:

    Outer loop ‚Üí for 3 floors.
    
    Inner loop ‚Üí for 4 rooms per floor.

## Invoice Tax Calculator
### Goal:

Take multiple items in multiple invoices and calculate tax per invoice.

### Steps:

Outer loop ‚Üí for 3 invoices.

Inner loop ‚Üí 4 items each.

Input item price, add to list.

Calculate total + 10% tax.

Print invoice total.

### Example:
    Invoice 1:
    Items: [100, 200, 300, 400]
    Invoice 2:
    Items: [250, 150, 350, 250]
    ‚Üí Totals: [1100, 1100]

```python
invoice_totals = []

for invoice_num in range(1, 4):
    print(f"Invoice {invoice_num}:")
    items = []
    for item_num in range(1, 5):
        while True:
            try:
                price = float(input(f"Item {item_num} price: "))
                if price >= 0:
                    break
                else:
                    print("Price cannot be negative.")
            except ValueError:
                print("Invalid input. Enter a number.")
        items.append(price)
    
    total = sum(items)
    total_with_tax = total * 1.1  # Adding 10% tax
    invoice_totals.append(total_with_tax)
    print(f"Invoice {invoice_num} total with tax: {total_with_tax}\n")

print("Invoice Totals:", invoice_totals)
```

## Student Result Sheet
### Goal:

    Generate report cards for each student in a class.

### Steps:

    Outer loop ‚Üí 5 students.
    
    Inner loop ‚Üí 4 subjects.

    If marks < 33 ‚Üí print ‚ÄúFailed in Subject‚Äù.

    Calculate total and grade per student.

### Example:
    Student 1: [80, 90, 70, 85] ‚Üí Grade: A  
    Student 2: [50, 32, 45, 40] ‚Üí Failed in Subject 2

```python
for student_num in range(1, 6):
    print(f"Enter marks for Student {student_num}:")
    marks = []
    failed = False
    
    for subject_num in range(1, 5):
        while True:
            try:
                mark = int(input(f"Subject {subject_num}: "))
                if 0 <= mark <= 100:
                    break
                else:
                    print("Marks must be between 0 and 100.")
            except ValueError:
                print("Invalid input. Enter a number.")
        marks.append(mark)
        if mark < 33:
            print(f"Failed in Subject {subject_num}")
            failed = True
    
    if not failed:
        total = sum(marks)
        avg = total / 4
        if avg >= 80:
            grade = "A"
        elif avg >= 60:
            grade = "B"
        elif avg >= 50:
            grade = "C"
        else:
            grade = "D"
        print(f"Student {student_num} Marks: {marks} ‚Üí Grade: {grade}\n")
    else:
        print(f"Student {student_num} Marks: {marks} ‚Üí Has failed in one or more subjects\n")
```

## Holiday Package Evaluator
### Goal:

Compare multiple travel packages across features like hotel, transport, food.

### Steps:

Outer loop ‚Üí 3 packages.

Inner loop ‚Üí 3 features.

Input rating out of 5.

Skip if rating = 0.

Print average rating per package.

### Example:
Package 1: [4, 5, 5] ‚Üí Avg = 4.67  
Package 2: [3, 4, 0] ‚Üí Avg = 3.5

```python
for package_num in range(1, 4):
    print(f"Enter ratings for Package {package_num}:")
    ratings = []
    
    for feature_num in range(1, 4):
        while True:
            try:
                rating = float(input(f"Feature {feature_num} rating (0-5): "))
                if 0 <= rating <= 5:
                    break
                else:
                    print("Rating must be between 0 and 5.")
            except ValueError:
                print("Invalid input. Enter a number.")
        
        if rating == 0:
            continue
        
        ratings.append(rating)
    
    avg_rating = sum(ratings) / len(ratings) if ratings else 0
    print(f"Package {package_num} Ratings: {ratings} ‚Üí Average = {avg_rating:.2f}\n")
```
